# SPCX Forecast — Keşif Notebook

Ana mantık `forecasting.py` modülünde. Bu notebook modülü kullanarak hızlı deney yapmak içindir.

> Yatırım tavsiyesi değildir; eğitim amaçlıdır.

```python
from forecasting import (
    DEFAULT_PARAMS,
    evaluate,
    fetch_data,
    forecast,
    naive_baseline_metrics,
    train_model,
)

TICKER = "SPCX"
HORIZON = 5
```

```python
df = fetch_data(TICKER)
df.tail()
```

```python
model = train_model(df, DEFAULT_PARAMS)
metrics = evaluate(model, df)
baseline = naive_baseline_metrics(df)

print(f"Model MAPE: {metrics['mape']:.2f}% | Naif MAPE: {baseline['mape']:.2f}%")
print(f"Model RMSE: ${metrics['rmse']:.2f} | Naif RMSE: ${baseline['rmse']:.2f}")
```

```python
pred = forecast(model, HORIZON, history_df=df)
pred[["ds", "yhat", "yhat_lower", "yhat_upper"]].tail(HORIZON + 2)
```

```python
from prophet.plot import plot_plotly

fig = plot_plotly(model, pred)
fig.show()
```
